```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from scipy import stats

# Load the training and testing data
X_train = np.load('X_train.npy')
X_test = np.load('X_test.npy')
y_train = np.load('y_train.npy')

# Convert X_train to a DataFrame and display summary statistics
X_train_df = pd.DataFrame(X_train)
print(X_train_df.describe())

# Uncomment if you want to inspect the summary statistics of y_train and X_test as well
# y_train_df = pd.DataFrame(y_train)
# print(y_train_df.describe())

# X_test_df = pd.DataFrame(X_test)
# print(X_test_df.describe())

# Calculate the Z-Score for each value in X_train
z_scores = np.abs(stats.zscore(X_train))

# Identify outliers where the Z-Score is greater than 3
outliers = (z_scores > 3).any(axis=1)

# Keep only the points that are not identified as outliers
X_train_clean = X_train[~outliers]
y_train_clean = y_train[~outliers]

# Convert the cleaned X_train to a DataFrame and display summary statistics
X_train_cleandf = pd.DataFrame(X_train_clean)
print(X_train_cleandf.describe())

# Add some Gaussian noise to y_train for experimental purposes
mean = 0
std = 0.01
noise = np.random.normal(mean, std, size=y_train.shape)

# Train a linear regression model on the training data
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Use the model to make predictions on the test data
y_test_pre = linear_model.predict(X_test)

# Save the predictions to a file
np.save('y_test.npy', y_test_pre)

# Using cross-validation to calculate the mean score (mean squared error)
scores = cross_val_score(linear_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
mean_score = -scores.mean()  # The result is negative, so invert it
print('Mean MSE from Cross-Validation:', mean_score)

# Define feature names (adjust as necessary)
features = ['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Feature 5']

# Choose the feature index to plot (3rd feature in this case, index 2)
feature_index = 2

# Plot the selected feature against y_train for both raw and cleaned data
plt.figure(figsize=(10, 6))
plt.scatter(X_train[:, feature_index], y_train, color='blue', label=f'Raw Data: {features[feature_index]} vs. Dependent Variable')
plt.scatter(X_train_clean[:, feature_index], y_train_clean, color='black', label=f'Cleaned Data: {features[feature_index]} vs. Dependent Variable', alpha=0.5)
plt.xlabel(features[feature_index])
plt.ylabel('Dependent Variable')
plt.title(f'Scatter Plot of {features[feature_index]} vs. Dependent Variable')
plt.legend()
plt.show()
```
